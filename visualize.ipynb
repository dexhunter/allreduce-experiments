````python
def read_algo(filename):
    x = []
    y = []

    with open(filename) as f:
        for line in f:
            num = line.split()
            x.append(num[0])
            y.append(num[1])

    x = np.array(x)
    y = np.array(y)

    return x.astype('float'), y.astype('float')

x, y1 = read_algo('data/2cpu/non-power-of-2/algo1.txt')
_, y2 = read_algo('data/2cpu/non-power-of-2/algo2.txt')
_, y3 = read_algo('data/2cpu/non-power-of-2/algo3.txt')
_, y4 = read_algo('data/2cpu/non-power-of-2/algo4.txt')
_, y5 = read_algo('data/2cpu/non-power-of-2/algo5.txt')
_, y6 = read_algo('data/2cpu/non-power-of-2/algo6.txt')

plt.title('Different Allreduce Algorithms among 2 CPUs (size increases non-power-of-two)')
plt.xlabel('message size(bytes)')
plt.ylabel('avg latency(ns)')

plt.plot(x, y1, label='algo1')
plt.plot(x, y2, label='algo2')
plt.plot(x, y3, label='algo3')
plt.plot(x, y4, label='algo4')
plt.plot(x, y5, label='algo5')
plt.plot(x, y6, label='algo6')

plt.legend()
````

````python
plt.show()
````

````python
df = pd.DataFrame({'x':x, 'basic_linear': y1, 'nonoverlapping': y2, 'recursive_doubling': y3, 'ring': y4, 'segmented_ring': y5, 'rabenseifner': y6})

# print(df)
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
plt.xlabel('message size(bytes)')
plt.ylabel('avg latency(ns)')
plt.title('Different Allreduce Algorithms among 2 CPUs (size increases non-power-of-two)')

# style
plt.style.use('seaborn-darkgrid')
 
# create a color palette
palette = plt.get_cmap('Set1')

# multiple line plot

num=0
for column in df.drop('x', axis=1):
    num+=1
    plt.plot(df['x'], df[column], marker='', color=palette(num), linewidth=1, alpha=0.9, label=column)

plt.xlim(9.70*1e8, 9.78*1e8)
plt.ylim(1.015*1e6, 1.04*1e6)

plt.legend()
````
